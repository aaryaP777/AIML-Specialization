```python
%matplotlib inline

import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

dataframe = pd.read_csv(r"C:\Users\aarya\Desktop\canada_per_capita_income.csv")
dataframe.head(10)

plt.xlabel("Year")
plt.ylabel("Per capita income")
plt.scatter(dataframe.year, dataframe.pci, color = 'blue', marker = 'o')

x = dataframe[["year"]]
y = dataframe["pci"]

model = linear_model.LinearRegression()
model.fit(x,y)

year_predict = model.predict(x)

plt.plot(dataframe.year, year_predict, color = "Red", linewidth = 2, label = "Best fit line")
```
