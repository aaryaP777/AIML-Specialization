```python
%matplotlib inline

import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

canada_df = pd.read_csv(r"C:\Users\aarya\Desktop\canada_per_capita_income.csv")
print(canada_df.head(10))
```

```python
plt.xlabel("Year")
plt.ylabel("Per capita income")
plt.scatter(canada_df.year, canada_df.pci, color = 'red', marker = 'o')
```

```python
year_input = canada_df.drop('pci', axis = 'columns')
year_input
```

```python
pci = canada_df.pci
pci
```

```python
reg = linear_model.LinearRegression()
reg.fit(year_input, pci)
```

```python
reg.predict(year_input)
```
