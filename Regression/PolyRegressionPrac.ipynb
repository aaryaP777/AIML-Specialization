```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
exp = np.array([0, 5, 10, 15, 20, 25, 30]).reshape(-1, 1)
salary = np.array([0, 1000, 2000, 3500, 4000, 4700, 6000])
```

```python
poly_small = PolynomialFeatures(degree = 2)
exp_poly_small = poly_small.fit_transform(exp)
linear_model_small = LinearRegression()
linear_model_small.fit(exp_poly_small, salary)
salary_poly_pred_small = linear_model_small.predict(exp_poly_small)
```

```python
plt.figure(figsize = (8, 5))
plt.scatter(exp, salary, color = 'green', label = 'Data', alpha = 0.8)
plt.plot(exp, salary_poly_pred_small, color = 'orange', label = 'polynomial reg', linewidth = 2)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
